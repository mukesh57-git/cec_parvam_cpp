## 1.1 Introduction to OOP

**Object-Oriented Programming (OOP)** is a programming paradigm that organizes software design around **objects** rather than **functions** and **logic**. An object is a self-contained entity that contains both **data** (attributes) and **behavior** (methods).

### What is OOP?

OOP models real-world entities as objects in code. For example, a `Car` object has:
- **Data** (attributes): `color`, `speed`, `fuelLevel`
- **Behavior** (methods): `accelerate()`, `brake()`, `refuel()`

### Why OOP?

| Problem with Procedural Programming | How OOP Solves It |
|---|---|
| Data and functions are separate → poor cohesion | Bundles data + behavior into objects |
| Global data is accessible everywhere → security issues | Encapsulation hides data |
| Code reuse requires copying functions | Inheritance reuses code cleanly |
| Large programs become spaghetti code | Modular class-based organization |
| Adding features breaks existing code | Polymorphism allows extensibility |

## 1.2 Procedural vs OOP

### Side-by-Side Comparison

| Aspect | Procedural (C) | OOP (C++) |
|---|---|---|
| **Focus** | Functions (what to do) | Objects (who does it) |
| **Data** | Global/local variables | Encapsulated in objects |
| **Code Organization** | Functions + structs | Classes + objects |
| **Reusability** | Copy-paste functions | Inheritance + composition |
| **Security** | Data accessible everywhere | Data hiding via access modifiers |
| **Real-world mapping** | Weak | Strong |
| **Maintainability** | Difficult for large projects | Easier with modular classes |

### Real-World Analogy

**Procedural Approach (C):**
You go to a restaurant. You tell the chef *how* to cook each step: "Chop vegetables, heat oil, add spices, stir..." — you control *every* detail.

**OOP Approach (C++):**
You order from a menu. The `Chef` object knows how to cook. You just call `chef.cookDish("Pasta")`. The `Chef` handles the details internally.

## 1.3 Four Pillars of OOP

### 1. **Class & Object**
- **Class**: A blueprint or template for creating objects. Defines what data and behavior all objects of that type will have.
- **Object**: A concrete instance of a class. It occupies memory and has its own copy of member variables.

**Real-world mapping:**
- Class = Blueprint of a house
- Object = The actual house built from that blueprint

### 2. **Encapsulation**
- Bundling **data** (member variables) and **behavior** (member functions) together inside a class.
- Hiding internal data from outside access using `private` access modifier.
- Exposing controlled access via `public` getter/setter functions.

### 3. **Inheritance**
- Creating new classes (`derived`) from existing ones (`base`).
- The derived class **reuses** the base class members and can add/extend functionality.
- Represents an **IS-A** relationship: "A Dog IS-A Animal"

### 4. **Polymorphism**
- **"One interface, many forms"**
- The ability to call the same function name but get different behaviors based on context.
- **Compile-time**: Function overloading, operator overloading
- **Runtime**: Virtual functions, overriding

### 5. **Abstraction**
- Hiding complex implementation details and showing only essential features.
- Achieved via **abstract classes** (with pure virtual functions) and **interfaces**.
- Example: You drive a car using the steering wheel — you don't need to know how the engine works internally.

## 1.4 Basic Structure of a C++ Program

```c++
// 1. Preprocessor directives — #include brings in header files
#include <iostream>   // For input/output (cin, cout)
#include <string>      // For string type

// 2. Using namespace (optional, but commonly used)
using namespace std;

// 3. Function prototype / Class definitions go here

// 4. Main function — entry point of every C++ program
int main() {
    // 5. Variable declarations
    // 6. Program logic (input, processing, output)
    // 7. Return statement
    return 0;
}
```

### Compilation Model Overview

```c++
Source Code (.cpp)          Header Files (.h / .hpp)
        |                          |
        ↓                          ↓
    Preprocessor — processes #include, #define, etc.
        |
        ↓
    Compiler — translates C++ code to object code (.o / .obj)
        |
        ↓
    Linker — combines object files + libraries into executable
        |
        ↓
    Executable (.exe / .out)
```

**Compilation Commands:**

```c++
# Compile a single file
g++ program.cpp -o program

# Compile multiple files
g++ main.cpp helper.cpp -o program

# Run the executable
./program        # Linux/Mac
program.exe      # Windows
```

## 1.5 Class and Object Syntax

### Declaring a Class

```c++
class ClassName {
    // Access specifier
    public:
        // Member variables (data)
        // Member functions (behavior)
    
    private:
        // Hidden data and helper functions
};  // <-- Note the semicolon!
```

### Creating Objects

```c++
ClassName objectName;                // Stack allocation
ClassName* ptr = new ClassName();    // Heap allocation (dynamic)
```

### Full Example

```c++
#include <iostream>
#include <string>
using namespace std;

// Class definition
class Student {
    public:
        // Member variables (attributes)
        string name;
        int age;
        double cgpa;
        
        // Member function (behavior)
        void displayInfo() {
            cout << "Name: " << name << endl;
            cout << "Age: " << age << endl;
            cout << "CGPA: " << cgpa << endl;
        }
};

int main() {
    // Creating an object on stack
    Student s1;
    s1.name = "Arjun";
    s1.age = 20;
    s1.cgpa = 8.75;
    s1.displayInfo();
    
    // Creating an object on heap
    Student* s2 = new Student();
    s2->name = "Priya";       // Arrow operator for pointer access
    s2->age = 21;
    s2->cgpa = 9.2;
    s2->displayInfo();
    
    delete s2;  // Free heap memory
    return 0;
}
```

**Output:**

```c++
Name: Arjun
Age: 20
CGPA: 8.75
Name: Priya
Age: 21
CGPA: 9.2
```

## 1.6 C++ vs C — Key Syntax Differences

| Feature | C | C++ |
|---|---|---|
| **Input** | `scanf("%d", &x);` | `cin >> x;` |
| **Output** | `printf("%d", x);` | `cout << x;` |
| **Function inside struct** | Not allowed | Allowed (struct acts like class) |
| **Default struct access** | `public` | `public` (struct) / `private` (class) |
| **Dynamic memory** | `malloc()` / `free()` | `new` / `delete` |
| **String handling** | `char[]` + `<string.h>` | `std::string` class |
| **Boolean type** | `int` (0/1) since C99 | `bool` (true/false) |
| **Code organization** | `.c` + `.h` files | `.cpp` + `.h` files, namespaces |

### Comparison Program

**C Style:**

```c++
#include <stdio.h>
#include <string.h>

struct Student {
    char name[50];
    int age;
};

int main() {
    struct Student s;
    strcpy(s.name, "Arjun");
    s.age = 20;
    printf("Name: %s, Age: %d\n", s.name, s.age);
    return 0;
}
```

**C++ Style:**

```c++
#include <iostream>
#include <string>
using namespace std;

class Student {
    public:
        string name;
        int age;
        void display() {
            cout << "Name: " << name << ", Age: " << age << endl;
        }
};

int main() {
    Student s;
    s.name = "Arjun";
    s.age = 20;
    s.display();
    return 0;
}
```

## Coding Examples

### Example 1 — Basic Class with Multiple Objects

```c++
#include <iostream>
#include <string>
using namespace std;

class Book {
    public:
        string title;
        string author;
        int pages;
        
        void readBook() {
            cout << "Reading \"" << title << "\" by " << author << endl;
        }
};

int main() {
    Book book1, book2;
    
    book1.title = "The Alchemist";
    book1.author = "Paulo Coelho";
    book1.pages = 197;
    
    book2.title = "1984";
    book2.author = "George Orwell";
    book2.pages = 328;
    
    book1.readBook();
    book2.readBook();
    
    return 0;
}
```

**Output:**

```c++
Reading "The Alchemist" by Paulo Coelho
Reading "1984" by George Orwell
```

### Example 2 — Using `cin` and `cout` with Objects

```c++
#include <iostream>
#include <string>
using namespace std;

class Rectangle {
    public:
        double length;
        double width;
        
        double area() {
            return length * width;
        }
        
        double perimeter() {
            return 2 * (length + width);
        }
};

int main() {
    Rectangle rect;
    
    cout << "Enter length: ";
    cin >> rect.length;
    cout << "Enter width: ";
    cin >> rect.width;
    
    cout << "\nRectangle Details:" << endl;
    cout << "Area      : " << rect.area() << endl;
    cout << "Perimeter : " << rect.perimeter() << endl;
    
    return 0;
}
```

**Sample Run:**

```c++
Enter length: 5.5
Enter width: 3.2

Rectangle Details:
Area      : 17.6
Perimeter : 17.4
```

### Example 3 — Multiple Objects from Different Classes

```c++
#include <iostream>
#include <string>
using namespace std;

class Engine {
    public:
        int horsepower;
        string fuelType;
        
        void start() {
            cout << horsepower << " HP " << fuelType << " engine started!" << endl;
        }
};

class Car {
    public:
        string brand;
        string model;
        Engine engine;  // Composition: Car HAS-A Engine
        
        void display() {
            cout << brand << " " << model << " with ";
            cout << engine.horsepower << " HP engine." << endl;
        }
};

int main() {
    Car myCar;
    myCar.brand = "Toyota";
    myCar.model = "Camry";
    myCar.engine.horsepower = 203;
    myCar.engine.fuelType = "Petrol";
    
    myCar.display();
    myCar.engine.start();
    
    return 0;
}
```

**Output:**

```c++
Toyota Camry with 203 HP engine.
203 HP Petrol engine started!
```

## Common Mistakes

| Mistake | Wrong | Correct |
|---|---|---|
| **Missing semicolon after class** | `class Car { ... }` | `class Car { ... };` |
| **Forgetting `#include <iostream>`** | Using `cout` without header | `#include <iostream>` at top |
| **Using `>>` wrong direction** | `cin << x;` | `cin >> x;` |
| **Missing `using namespace std;`** | `cout << "Hi";` → error | Add `using namespace std;` or use `std::cout` |
| **Forgetting `return 0;`** | Main function missing return | `return 0;` at end of `main()` |
| **Dot vs Arrow confusion** | `ptr.name` for pointer | `ptr->name` or `(*ptr).name` |
| **Struct vs Class default access** | Expecting struct members to be private | Struct default is `public`; class default is `private` |

## Interview Questions

1. **What is the difference between a class and an object?**
   - A class is a blueprint/template. An object is a concrete instance of a class that occupies memory. A class defines structure; an object holds actual data.

2. **Explain the four pillars of OOP.**
   - **Encapsulation**: Bundling data + methods, hiding internal state
   - **Inheritance**: Reusing code via parent-child relationships
   - **Polymorphism**: Same interface, different behaviors
   - **Abstraction**: Hiding complexity, showing only essentials

3. **What is the difference between C and C++?**
   - C is procedural; C++ supports OOP. C++ has classes, objects, inheritance, polymorphism, exceptions, templates, STL, references, and `new`/`delete`.

4. **What is the default access level in a C++ class vs struct?**
   - Class: `private` by default. Struct: `public` by default.

5. **Explain the compilation process of a C++ program.**
   - **Preprocessing**: Handles `#include`, `#define`, macros
   - **Compilation**: Translates C++ to object code (`.o`/`.obj`)
   - **Linking**: Combines object files + libraries into executable

## Quick Revision Summary

- **OOP** = Organizing code around objects (data + behavior)
- **Four pillars**: Encapsulation, Inheritance, Polymorphism, Abstraction
- **Class** = Blueprint; **Object** = Instance
- **`class`** defaults to `private` access; **`struct`** defaults to `public`
- **Stack objects**: `ClassName obj;` — use dot operator (`.`)
- **Heap objects**: `ClassName* ptr = new ClassName();` — use arrow operator (`->`)

- **Files**: `.cpp` for implementation, `.h` / `.hpp` for headers
