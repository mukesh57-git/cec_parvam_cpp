## 2.1 Defining and Declaring Classes

### Class Definition Syntax

```c++
class ClassName {
    // Access specifier (default is private for class)
    private:
        // Private members — accessible only within the class
        
    public:
        // Public members — accessible from anywhere
        
    protected:
        // Protected members — accessible in class + derived classes
};  // <-- Semicolon is mandatory
```

### Separating Declaration (.h) from Definition (.cpp)

For larger projects, it's best practice to separate the class **declaration** (header file) from the **implementation** (source file).

**Student.h** — Header file (declaration)

```c++
#ifndef STUDENT_H   // Include guard — prevents multiple inclusions
#define STUDENT_H

#include <string>
using namespace std;

class Student {
    private:
        string name;
        int age;
        double cgpa;
    
    public:
        // Constructor declarations
        Student();
        Student(string n, int a, double c);
        
        // Method declarations
        void setName(string n);
        string getName();
        void displayInfo();
};

#endif
```

**Student.cpp** — Source file (implementation)

```c++
#include "Student.h"
#include <iostream>
using namespace std;

// Constructor definitions
Student::Student() {
    name = "Unknown";
    age = 0;
    cgpa = 0.0;
}

Student::Student(string n, int a, double c) {
    name = n;
    age = a;
    cgpa = c;
}

// Method definitions
void Student::setName(string n) {
    name = n;
}

string Student::getName() {
    return name;
}

void Student::displayInfo() {
    cout << "Name: " << name << ", Age: " << age << ", CGPA: " << cgpa << endl;
}
```

**main.cpp** — Using the class

```c++
#include "Student.h"

int main() {
    Student s1("Arjun", 20, 8.75);
    s1.displayInfo();
    return 0;
}
```

### Include Guards Explained

```c++
#ifndef HEADER_NAME_H  // If not already defined...
#define HEADER_NAME_H  // ...define it now
    // ... header content ...
#endif                 // End of conditional
```

This prevents the same header from being included multiple times in a single compilation unit, which would cause "redefinition" errors.

## 2.2 Access Modifiers

| Modifier | Access Level | Within Class | Derived Class | Outside Class |
|---|---|---|---|---|
| `public` | Most accessible | ✅ | ✅ | ✅ |
| `protected` | Limited | ✅ | ✅ | ❌ |
| `private` | Most restricted | ✅ | ❌ | ❌ |

### Default Access: Class vs Struct

```c++
class MyClass {
    // DEFAULT: private
    int x;  // private by default
};

struct MyStruct {
    // DEFAULT: public
    int y;  // public by default
};
```

**Key difference:** In a `class`, members are `private` by default. In a `struct`, members are `public` by default. Otherwise, they are functionally identical in C++.

### Full Access Modifiers Example

```c++
#include <iostream>
using namespace std;

class BankAccount {
    private:    // Hidden from outside
        double balance;
        string accountNumber;
    
    protected:  // Accessible in derived classes
        string accountHolderName;
    
    public:     // Public interface
        BankAccount(string name, string accNo) {
            accountHolderName = name;
            accountNumber = accNo;
            balance = 0.0;
        }
        
        void deposit(double amount) {
            if (amount > 0) {
                balance += amount;
                cout << "Deposited $" << amount << endl;
            }
        }
        
        void showBalance() {
            cout << "Balance: $" << balance << endl;
        }
};

class SavingsAccount : public BankAccount {
    public:
        SavingsAccount(string name, string accNo) 
            : BankAccount(name, accNo) {}
        
        void showHolder() {
            // Can access protected member from base class
            cout << "Account holder: " << accountHolderName << endl;
            
            // Cannot access private member balance directly
            // cout << balance;  // ERROR! balance is private
        }
};

int main() {
    BankAccount acc("Arjun", "A1001");
    // acc.balance = 5000;        // ERROR! private
    // acc.accountNumber = "X";   // ERROR! private
    
    acc.deposit(1000);
    acc.showBalance();
    
    SavingsAccount sav("Priya", "S2001");
    sav.showHolder();
    
    return 0;
}
```

**Output:**

```c++
Deposited $1000
Balance: $1000
Account holder: Priya
```

## 2.3 Object Instantiation

### Stack vs Heap Objects

| Aspect | Stack Object | Heap Object |
|---|---|---|
| **Declaration** | `Student s;` | `Student* s = new Student();` |
| **Memory allocation** | Automatic | Manual (`new`) |
| **Deallocation** | Automatic (out of scope) | Manual (`delete`) |
| **Member access** | `s.name` (dot) | `s->name` (arrow) or `(*s).name` |
| **Array syntax** | `Student arr[10];` | `Student* arr = new Student[10];` |
| **Size limits** | Smaller (typically ~1-8 MB) | Larger (heap is GBs) |
| **Speed** | Faster | Slower (allocation overhead) |

### Dot Operator (.) for Member Access

Used when you have an **object directly** (not a pointer).

```c++
Student s1;          // Stack object
s1.name = "Arjun";   // Dot operator
s1.displayInfo();    // Dot operator to call method
```

### Arrow Operator (->) for Pointer Access

Used when you have a **pointer to an object**.

```c++
Student* s2 = new Student();  // Heap object — returns pointer
s2->name = "Priya";           // Arrow operator (dereferences + accesses)
s2->displayInfo();            // Equivalent to: (*s2).displayInfo();

// You can also use dot operator with explicit dereference:
(*s2).name = "Priya";         // Same as s2->name
```

## 2.4 Scope Resolution Operator (::)

The `::` operator is used to:
1. Define member functions outside the class
2. Access global variables when shadowed by local/class variables
3. Access static members of a class
4. Access names in a namespace

### Defining Member Functions Outside the Class

```c++
class Example {
    public:
        void display();     // Declaration only
};

// Definition using scope resolution
void Example::display() {   // ClassName::FunctionName
    cout << "Hello from Example!" << endl;
}
```

### Accessing Global vs Class Scope

```c++
#include <iostream>
using namespace std;

int value = 100;  // Global variable

class Demo {
    public:
        int value;  // Member variable
        
        void setValue(int value) {        // Parameter shadows member
            this->value = value;           // 'this->' accesses member
        }
        
        void printAll() {
            int value = 50;                // Local variable
            cout << "Local   value: " << value << endl;
            cout << "Member  value: " << this->value << endl;
            cout << "Global  value: " << ::value << endl;  // :: global scope
        }
};

int main() {
    Demo d;
    d.setValue(200);
    d.printAll();
    return 0;
}
```

**Output:**

```c++
Local   value: 50
Member  value: 200
Global  value: 100
```

### Using :: with Static Members

```c++
class Config {
    public:
        static string appName;    // Declaration
        static void showVersion() { cout << "v1.0" << endl; }
};

// Definition of static member (outside class)
string Config::appName = "MyApp";

int main() {
    // Access via class name (no object needed)
    cout << Config::appName << endl;
    Config::showVersion();
    return 0;
}
```

## Coding Examples

### Example 1 — Full Class with Separation of Concerns

**Rectangle.h**

```c++
#ifndef RECTANGLE_H
#define RECTANGLE_H

class Rectangle {
    private:
        double length;
        double width;
    
    public:
        Rectangle();                    // Default constructor
        Rectangle(double l, double w); // Parameterized constructor
        void setDimensions(double l, double w);
        double area();
        double perimeter();
        void display();
};

#endif
```

**Rectangle.cpp**

```c++
#include "Rectangle.h"
#include <iostream>
using namespace std;

Rectangle::Rectangle() {
    length = 0;
    width = 0;
}

Rectangle::Rectangle(double l, double w) {
    length = l;
    width = w;
}

void Rectangle::setDimensions(double l, double w) {
    length = l;
    width = w;
}

double Rectangle::area() {
    return length * width;
}

double Rectangle::perimeter() {
    return 2 * (length + width);
}

void Rectangle::display() {
    cout << "Rectangle: " << length << " x " << width << endl;
    cout << "Area: " << area() << ", Perimeter: " << perimeter() << endl;
}
```

**main.cpp**

```c++
#include "Rectangle.h"

int main() {
    Rectangle r1;                     // Default constructor
    Rectangle r2(5.0, 3.0);          // Parameterized constructor
    
    r1.display();                     // 0 x 0
    r2.display();                     // 5 x 3
    
    r1.setDimensions(4.5, 2.5);
    r1.display();                     // 4.5 x 2.5
    
    return 0;
}
```

**Output:**

```c++
Rectangle: 0 x 0
Area: 0, Perimeter: 0
Rectangle: 5 x 3
Area: 15, Perimeter: 16
Rectangle: 4.5 x 2.5
Area: 11.25, Perimeter: 14
```

### Example 2 — Stack vs Heap Object Demonstration

```c++
#include <iostream>
using namespace std;

class Demo {
    public:
        int id;
        Demo(int i) {
            id = i;
            cout << "Constructor called for object " << id << endl;
        }
        ~Demo() {
            cout << "Destructor called for object " << id << endl;
        }
};

int main() {
    cout << "--- Stack object ---" << endl;
    Demo stackObj(1);          // Created on stack
    
    cout << "\n--- Heap object ---" << endl;
    Demo* heapObj = new Demo(2);  // Created on heap
    
    cout << "\nStack object goes out of scope automatically..." << endl;
    // stackObj automatically destroyed when main() ends
    
    cout << "Heap object must be explicitly deleted..." << endl;
    delete heapObj;  // Manual deletion
    
    cout << "\nEnd of main()" << endl;
    return 0;
}
```

**Output:**

```c++
--- Stack object ---
Constructor called for object 1

--- Heap object ---
Constructor called for object 2

Stack object goes out of scope automatically...
Heap object must be explicitly deleted...
Destructor called for object 2

End of main()
Destructor called for object 1
```

### Example 3 — Array of Objects

```c++
#include <iostream>
#include <string>
using namespace std;

class Student {
    public:
        string name;
        int marks;
        
        Student() { name = "Unknown"; marks = 0; }
        Student(string n, int m) { name = n; marks = m; }
        
        void display() {
            cout << name << ": " << marks << endl;
        }
};

int main() {
    // Array of objects on stack
    Student batch1[3] = {
        Student("Arjun", 85),
        Student("Priya", 92),
        Student("Rahul", 78)
    };
    
    cout << "Batch 1 (Stack array):" << endl;
    for (int i = 0; i < 3; i++) {
        batch1[i].display();
    }
    
    // Dynamic array of objects on heap
    Student* batch2 = new Student[3] {
        Student("Sneha", 88),
        Student("Vikram", 76),
        Student("Ananya", 95)
    };
    
    cout << "\nBatch 2 (Heap array):" << endl;
    for (int i = 0; i < 3; i++) {
        batch2[i].display();  // Using dot operator with pointer-as-array
    }
    
    delete[] batch2;  // Must use delete[] for arrays
    return 0;
}
```

**Output:**

```c++
Batch 1 (Stack array):
Arjun: 85
Priya: 92
Rahul: 78

Batch 2 (Heap array):
Sneha: 88
Vikram: 76
Ananya: 95
```

## Common Mistakes

| Mistake | Wrong | Correct |
|---|---|---|
| **Missing semicolon after class** | `class A { }` | `class A { };` |
| **Forgetting include guard** | No `#ifndef` leads to redefinition errors | Use `#ifndef HEADER_H` / `#define HEADER_H` |
| **Using dot on pointer** | `ptr.name` | `ptr->name` or `(*ptr).name` |
| **Forgetting `delete` for heap** | `new` without `delete` = memory leak | Always pair `new` with `delete` |
| **Using `delete` instead of `delete[]`** | `delete arr;` for array | `delete[] arr;` |
| **Accessing private members** | `obj.privateVar` from outside | Use public getter/setter |
| **Confusing class vs struct access** | `class A { int x; };` — x is **private** by default | Struct is `public` by default |

## Interview Questions

1. **What is the difference between a class and a struct in C++?**
   - Default access: class members are `private`, struct members are `public`. Otherwise, they are identical in C++ (unlike C).

2. **Explain stack vs heap allocation of objects.**
   - Stack: automatic allocation/deallocation, faster, limited size. Heap: manual (`new`/`delete`), slower, larger capacity.

3. **What does the scope resolution operator `::` do?**
   - It defines the scope context. Used for: defining member functions outside class, accessing global variables, accessing static members, specifying namespaces.

4. **Why use header files and separate .cpp files?**
   - Separation of interface (`.h`) from implementation (`.cpp`) improves compilation time, code organization, and encapsulation. Changes to implementation don't require recompiling other files that include the header.

5. **What is an include guard and why is it important?**
   - `#ifndef` / `#define` / `#endif` prevents a header from being included multiple times in one compilation unit, which would cause "redefinition" errors.

## Quick Revision Summary

- **Class declaration** in `.h`; **definition** in `.cpp` for large projects
- **Include guards** (`#ifndef`/`#define`/`#endif`) prevent double inclusion
- **`public`**: accessible everywhere; **`private`**: within class only; **`protected`**: class + derived
- **Class default** = `private`; **Struct default** = `public`
- **Stack**: `Student s;` — dot operator (`.`)
- **Heap**: `Student* s = new Student();` — arrow operator (`->`)
- **`::`** = scope resolution — defines which class/namespace a member belongs to
- **Always pair** `new` with `delete`, and `new[]` with `delete[]`
