## 5.1 Introduction to Inheritance

**Inheritance** allows a class (derived/child class) to **inherit** properties and behaviors from another class (base/parent class). It represents an **IS-A** relationship.

### Basic Syntax

```c++
class BaseClass {
    // Base class members
};

class DerivedClass : accessSpecifier BaseClass {
    // Additional members specific to derived class
};
```

### Real-World Analogy

```c++
Vehicle (Base Class)
         /         |         \
        /          |          \
     Car          Bike        Truck  (Derived Classes)
    - 4 wheels   - 2 wheels   - 6 wheels
    - 5 seats    - 2 seats    - 2 seats
       |            |
   Sedan        Sports Bike
   SUV          Cruiser
```

Every `Car` **IS-A** `Vehicle`. Every `Sedan` **IS-A** `Car` **AND** IS-A `Vehicle`.

### Basic Inheritance Example

```c++
#include <iostream>
#include <string>
using namespace std;

// Base class
class Animal {
    protected:
        string name;
        int age;
    
    public:
        Animal(string n, int a) : name(n), age(a) {}
        
        void eat() {
            cout << name << " is eating." << endl;
        }
        
        void sleep() {
            cout << name << " is sleeping." << endl;
        }
};

// Derived class
class Dog : public Animal {
    private:
        string breed;
    
    public:
        // Constructor calls base class constructor
        Dog(string n, int a, string b) : Animal(n, a), breed(b) {}
        
        void bark() {
            cout << name << " (the " << breed << ") is barking!" << endl;
        }
        
        void display() {
            cout << "Dog: " << name << ", Age: " << age << ", Breed: " << breed << endl;
        }
};

int main() {
    Dog myDog("Buddy", 3, "Golden Retriever");
    
    myDog.eat();      // Inherited from Animal
    myDog.sleep();    // Inherited from Animal
    myDog.bark();     // Own method
    myDog.display();
    
    return 0;
}
```

**Output:**

```c++
Buddy is eating.
Buddy is sleeping.
Buddy (the Golden Retriever) is barking!
Dog: Buddy, Age: 3, Breed: Golden Retriever
```

### public, protected, private Inheritance

The access specifier in inheritance determines how the base class members are **inherited** into the derived class.

| Base Member Access | `public` Inheritance | `protected` Inheritance | `private` Inheritance |
|---|---|---|---|
| `public` | → `public` in derived | → `protected` in derived | → `private` in derived |
| `protected` | → `protected` in derived | → `protected` in derived | → `private` in derived |
| `private` | → hidden (not accessible) | → hidden (not accessible) | → hidden (not accessible) |

```c++
class Base {
    public:
        int pub;
    protected:
        int prot;
    private:
        int priv;
};

class PubDerived : public Base {
    // pub → public
    // prot → protected
    // priv → not accessible
    void example() {
        pub = 1;    // ✅ OK
        prot = 2;   // ✅ OK (protected in derived)
        // priv = 3; // ❌ ERROR: not accessible
    }
};

class ProtDerived : protected Base {
    // pub → protected
    // prot → protected
    // priv → not accessible
    void example() {
        pub = 1;    // ✅ OK (now protected)
        prot = 2;   // ✅ OK
    }
};

class PrivDerived : private Base {
    // pub → private
    // prot → private
    // priv → not accessible
    void example() {
        pub = 1;    // ✅ OK (now private)
        prot = 2;   // ✅ OK
    }
};

int main() {
    Base b;
    b.pub = 10;     // ✅ OK (public)
    // b.prot = 20; // ❌ ERROR (protected)
    
    PubDerived pd;
    pd.pub = 10;    // ✅ OK (still public)
    
    ProtDerived prd;
    // prd.pub = 10; // ❌ ERROR (now protected)
    
    PrivDeriveD prvd;
    // prvd.pub = 10; // ❌ ERROR (now private)
    
    return 0;
}
```

> **Rule of thumb:** Use `public` inheritance for IS-A relationships. `protected` and `private` inheritance are rarely used (mainly for implementation reuse).

## 5.2 Types of Inheritance

### 1. Single Inheritance

One derived class inherits from one base class.

```c++
class A { /* ... */ };
class B : public A { /* ... */ };
```

```c++
A
    |
    B
```

### 2. Multiple Inheritance

One derived class inherits from **multiple** base classes.

```c++
class A { /* ... */ };
class B { /* ... */ };
class C : public A, public B { /* ... */ };
```

```c++
A    B
     \  /
      C
```

### 3. Multilevel Inheritance

A class inherits from a derived class (chain).

```c++
class A { /* ... */ };
class B : public A { /* ... */ };
class C : public B { /* ... */ };
```

```c++
A
    |
    B
    |
    C
```

### 4. Hierarchical Inheritance

Multiple classes derive from a single base class.

```c++
class A { /* ... */ };
class B : public A { /* ... */ };
class C : public A { /* ... */ };
class D : public A { /* ... */ };
```

```c++
A
     /  |  \
    B   C   D
```

### 5. Hybrid (Virtual) Inheritance

Combination of multiple + multilevel inheritance. Uses **virtual base classes** to solve the **diamond problem**.

```c++
class A { /* ... */ };
class B : virtual public A { /* ... */ };
class C : virtual public A { /* ... */ };
class D : public B, public C { /* ... */ };
```

```c++
A
     / \
    B   C    (both virtual)
     \ /
      D
```

### Full Multi-Type Inheritance Example

```c++
#include <iostream>
#include <string>
using namespace std;

// ========= SINGLE INHERITANCE =========
class Person {
    protected:
        string name;
        int age;
    
    public:
        Person(string n, int a) : name(n), age(a) {}
        void display() {
            cout << "Name: " << name << ", Age: " << age;
        }
};

class Student : public Person {
    private:
        int rollNo;
    
    public:
        Student(string n, int a, int r) : Person(n, a), rollNo(r) {}
        
        void display() {
            Person::display();
            cout << ", Roll No: " << rollNo << endl;
        }
};

// ========= MULTILEVEL INHERITANCE =========
class GraduateStudent : public Student {
    private:
        string thesisTopic;
    
    public:
        GraduateStudent(string n, int a, int r, string t) 
            : Student(n, a, r), thesisTopic(t) {}
        
        void display() {
            Student::display();
            cout << "Thesis: " << thesisTopic << endl;
        }
};

// ========= HIERARCHICAL INHERITANCE =========
class Teacher : public Person {
    private:
        double salary;
    
    public:
        Teacher(string n, int a, double s) : Person(n, a), salary(s) {}
        
        void display() {
            Person::display();
            cout << ", Salary: $" << salary << endl;
        }
};

// ========= MULTIPLE INHERITANCE =========
class TeachingAssistant : public Student, public Teacher {
    public:
        TeachingAssistant(string n, int a, int r, double s) 
            : Student(n, a, r), Teacher(n, a, s) {}
        
        void display() {
            cout << "Teaching Assistant: " << name << endl;
            cout << "Age: " << age << ", Roll: " << rollNo << endl;
        }
};

int main() {
    cout << "=== Single Inheritance ===" << endl;
    Student s1("Arjun", 20, 101);
    s1.display();
    
    cout << "\n=== Multilevel Inheritance ===" << endl;
    GraduateStudent gs("Priya", 24, 201, "Machine Learning");
    gs.display();
    
    cout << "\n=== Hierarchical Inheritance ===" << endl;
    Teacher t1("Dr. Sharma", 45, 80000);
    t1.display();
    
    return 0;
}
```

**Output:**

```c++
=== Single Inheritance ===
Name: Arjun, Age: 20, Roll No: 101

=== Multilevel Inheritance ===
Name: Priya, Age: 24, Roll No: 201
Thesis: Machine Learning

=== Hierarchical Inheritance ===
Name: Dr. Sharma, Age: 45, Salary: $80000
```

### The Diamond Problem (Virtual Inheritance)

```c++
#include <iostream>
using namespace std;

// Base class
class Animal {
    protected:
        int age;
    
    public:
        Animal() : age(0) {
            cout << "Animal constructor" << endl;
        }
};

// Without virtual inheritance — each has its own Animal
class LandAnimal : virtual public Animal {
    public:
        LandAnimal() { cout << "LandAnimal constructor" << endl; }
};

class WaterAnimal : virtual public Animal {
    public:
        WaterAnimal() { cout << "WaterAnimal constructor" << endl; }
};

// With virtual inheritance — only ONE Animal instance
class Amphibian : public LandAnimal, public WaterAnimal {
    public:
        Amphibian() { cout << "Amphibian constructor" << endl; }
        
        void setAge(int a) { age = a; }     // No ambiguity! Only one 'age'
        int getAge() { return age; }
};

int main() {
    Amphibian frog;
    frog.setAge(5);
    cout << "Age: " << frog.getAge() << endl;
    
    cout << "\nNote: Animal constructor called only ONCE!" << endl;
    return 0;
}
```

**Output:**

```c++
Animal constructor
LandAnimal constructor
WaterAnimal constructor
Amphibian constructor
Age: 5

Note: Animal constructor called only ONCE!
```

> **Without virtual inheritance:** `Amphibian` would have **two** copies of `Animal` members, causing ambiguity. `virtual` ensures only **one** shared base class instance.

## 5.3 Constructor Execution Order

### Order of Construction

When a derived class object is created, constructors are called in this order:

1. **Base class constructor** (base-to-derived order in multiple inheritance)
2. **Member objects' constructors** (in order of declaration)
3. **Derived class constructor body**

```c++
#include <iostream>
using namespace std;

class A {
    public:
        A() { cout << "A constructor" << endl; }
        ~A() { cout << "A destructor" << endl; }
};

class B : public A {
    public:
        B() { cout << "B constructor" << endl; }
        ~B() { cout << "B destructor" << endl; }
};

class C : public B {
    public:
        C() { cout << "C constructor" << endl; }
        ~C() { cout << "C destructor" << endl; }
};

int main() {
    cout << "Creating C object:" << endl;
    C obj;
    cout << "\nDestroying C object:" << endl;
    return 0;
}
```

**Output:**

```c++
Creating C object:
A constructor
B constructor
C constructor

Destroying C object:
C destructor
B destructor
A destructor
```

> **Destructors fire in reverse order** — like a stack (last constructed, first destroyed).

### Passing Arguments to Base Constructor

```c++
#include <iostream>
#include <string>
using namespace std;

class Vehicle {
    protected:
        string brand;
        int year;
    
    public:
        Vehicle(string b, int y) : brand(b), year(y) {
            cout << "Vehicle constructor: " << brand << " (" << year << ")" << endl;
        }
        ~Vehicle() {
            cout << "Vehicle destructor: " << brand << endl;
        }
};

class Car : public Vehicle {
    private:
        string model;
    
    public:
        // Pass arguments to base class constructor in initialization list
        Car(string b, int y, string m) : Vehicle(b, y), model(m) {
            cout << "Car constructor: " << brand << " " << model << endl;
        }
        ~Car() {
            cout << "Car destructor: " << model << endl;
        }
        
        void display() {
            cout << year << " " << brand << " " << model << endl;
        }
};

int main() {
    Car myCar("Toyota", 2024, "Camry");
    myCar.display();
    cout << endl;
    return 0;
}
```

**Output:**

```c++
Vehicle constructor: Toyota (2024)
Car constructor: Toyota Camry
2024 Toyota Camry

Car destructor: Camry
Vehicle destructor: Toyota
```

## Coding Examples

### Example 1 — Employee Hierarchy

```c++
#include <iostream>
#include <string>
using namespace std;

// Base class
class Employee {
    protected:
        int id;
        string name;
        double baseSalary;
    
    public:
        Employee(int i, string n, double s) 
            : id(i), name(n), baseSalary(s) {}
        
        virtual double calculateSalary() {
            return baseSalary;
        }
        
        void display() {
            cout << "ID: " << id << ", Name: " << name;
            cout << ", Salary: $" << calculateSalary() << endl;
        }
        
        virtual ~Employee() {}
};

// Derived class 1
class Manager : public Employee {
    private:
        double bonus;
    
    public:
        Manager(int i, string n, double s, double b) 
            : Employee(i, n, s), bonus(b) {}
        
        double calculateSalary() override {
            return baseSalary + bonus;
        }
};

// Derived class 2
class Intern : public Employee {
    private:
        int stipend;
    
    public:
        Intern(int i, string n, double s, int st) 
            : Employee(i, n, s), stipend(st) {}
        
        double calculateSalary() override {
            return stipend;  // Interns get fixed stipend
        }
};

int main() {
    Employee* employees[3];
    
    employees[0] = new Employee(101, "Arjun", 50000);
    employees[1] = new Manager(102, "Priya", 80000, 15000);
    employees[2] = new Intern(103, "Rahul", 0, 15000);
    
    for (int i = 0; i < 3; i++) {
        employees[i]->display();
    }
    
    for (int i = 0; i < 3; i++) {
        delete employees[i];
    }
    
    return 0;
}
```

**Output:**

```c++
ID: 101, Name: Arjun, Salary: $50000
ID: 102, Name: Priya, Salary: $95000
ID: 103, Name: Rahul, Salary: $15000
```

### Example 2 — Multiple Inheritance (Diamond Problem Solution)

```c++
#include <iostream>
#include <string>
using namespace std;

class Person {
    protected:
        string name;
    
    public:
        Person(string n) : name(n) {
            cout << "Person created: " << name << endl;
        }
        virtual ~Person() {}
};

class Employee : virtual public Person {
    protected:
        int employeeId;
    
    public:
        Employee(string n, int id) : Person(n), employeeId(id) {
            cout << "Employee created: " << name << " (ID: " << id << ")" << endl;
        }
};

class Student : virtual public Person {
    protected:
        int studentId;
    
    public:
        Student(string n, int id) : Person(n), studentId(id) {
            cout << "Student created: " << name << " (ID: " << id << ")" << endl;
        }
};

// Intern inherits from both — only ONE Person instance due to virtual
class Intern : public Employee, public Student {
    private:
        string project;
    
    public:
        Intern(string n, int empId, int stuId, string proj) 
            : Person(n), Employee(n, empId), Student(n, stuId), project(proj) {
            cout << "Intern created: " << name << " on project " << project << endl;
        }
        
        void display() {
            cout << "\nIntern Details:" << endl;
            cout << "Name     : " << name << endl;        // No ambiguity!
            cout << "Emp ID   : " << employeeId << endl;
            cout << "Stud ID  : " << studentId << endl;
            cout << "Project  : " << project << endl;
        }
};

int main() {
    Intern intern("Arjun", 1001, 5001, "AI Research");
    intern.display();
    return 0;
}
```

**Output:**

```c++
Person created: Arjun
Employee created: Arjun (ID: 1001)
Student created: Arjun (ID: 5001)
Intern created: Arjun on project AI Research

Intern Details:
Name     : Arjun
Emp ID   : 1001
Stud ID  : 5001
Project  : AI Research
```

### Example 3 — Vehicle Rental System (Multilevel Inheritance)

A real-world vehicle rental system demonstrating multilevel and hierarchical inheritance.

```c++
#include <iostream>
#include <string>
using namespace std;

// Base class
class Vehicle {
    protected:
        string regNumber;
        double baseRate;
    
    public:
        Vehicle(string reg, double rate) : regNumber(reg), baseRate(rate) {}
        
        virtual double calculateRental(int days) {
            return baseRate * days;
        }
        
        void display() {
            cout << "Vehicle: " << regNumber << ", Rate: $" << baseRate << "/day" << endl;
        }
        
        virtual ~Vehicle() {}
};

// Derived class - Car (adds AC charge)
class Car : public Vehicle {
    protected:
        bool hasAC;
    
    public:
        Car(string reg, double rate, bool ac) 
            : Vehicle(reg, rate), hasAC(ac) {}
        
        double calculateRental(int days) override {
            double total = baseRate * days;
            if (hasAC) total += 200 * days;  // Extra AC charge
            return total;
        }
};

// Further derived - SUV (adds luggage capacity surcharge)
class SUV : public Car {
    private:
        int luggageCapacity;
    
    public:
        SUV(string reg, double rate, bool ac, int luggage) 
            : Car(reg, rate, ac), luggageCapacity(luggage) {}
        
        double calculateRental(int days) override {
            double total = Car::calculateRental(days);
            if (luggageCapacity > 400) total += 500 * days;  // Large luggage surcharge
            return total;
        }
};

// Another derived class - Bike
class Bike : public Vehicle {
    private:
        int engineCC;
    
    public:
        Bike(string reg, double rate, int cc) 
            : Vehicle(reg, rate), engineCC(cc) {}
        
        double calculateRental(int days) override {
            double total = baseRate * days;
            if (engineCC > 200) total *= 1.15;  // Higher CC bikes cost more
            return total;
        }
};

int main() {
    Vehicle* fleet[3];
    
    fleet[0] = new Car("KA-01-AB-1234", 1500, true);
    fleet[1] = new SUV("KA-02-XY-5678", 2500, true, 500);
    fleet[2] = new Bike("KA-03-PQ-9012", 800, 350);
    
    int rentalDays = 5;
    cout << "=== Vehicle Rental Report (" << rentalDays << " days) ===" << endl;
    cout << "--------------------------------------------" << endl;
    
    for (int i = 0; i < 3; i++) {
        fleet[i]->display();
        cout << "  Rental Cost: $" << fleet[i]->calculateRental(rentalDays) << endl;
        cout << "--------------------------------------------" << endl;
    }
    
    for (int i = 0; i < 3; i++) delete fleet[i];
    return 0;
}
```

**Output:**

```c++
=== Vehicle Rental Report (5 days) ===
--------------------------------------------
Vehicle: KA-01-AB-1234, Rate: $1500/day
  Rental Cost: $8500
--------------------------------------------
Vehicle: KA-02-XY-5678, Rate: $2500/day
  Rental Cost: $15000
--------------------------------------------
Vehicle: KA-03-PQ-9012, Rate: $800/day
  Rental Cost: $4600
--------------------------------------------
```

## Common Mistakes

| Mistake | Wrong | Correct |
|---|---|---|
| **Forgetting to call base constructor** | `Derived(int x, int y) { }` — base not initialized | `Derived(int x, int y) : Base(x) { }` |
| **Private inheritance thinking it's IS-A** | `class B : private A` — not a true IS-A | Use `public` inheritance for IS-A |
| **Diamond problem without virtual** | Two copies of base class members | Use `virtual` inheritance |
| **Forgetting virtual destructor** | Deleting derived via base pointer — undefined behavior | `virtual ~Base() {}` |
| **Trying to access base private members** | `this->privateMember` in derived | Only `protected` and `public` are accessible |
| **Call order confusion** | Thinking derived constructor runs first | Base constructor ALWAYS runs first |

## Interview Questions

1. **What is inheritance and what are its types?**
   - Inheritance allows a class to reuse properties of another class. Types: **Single**, **Multiple**, **Multilevel**, **Hierarchical**, **Hybrid** (virtual).

2. **What is the diamond problem and how is it solved?**
   - When a class inherits from two classes that share a common base, causing ambiguity. Solved using **virtual inheritance** (`class B : virtual public A`).

3. **What is the difference between public, protected, and private inheritance?**
   - Controls how base class members are inherited: `public` preserves access levels, `protected` makes public→protected, `private` makes everything private in derived.

4. **In what order are constructors and destructors called in inheritance?**
   - **Constructors**: Base → Derived (top-down). **Destructors**: Derived → Base (bottom-up, reverse order).

5. **Can a derived class access private members of its base class?**
   - No. Derived classes can only access `public` and `protected` members of the base class. Private members are hidden.

## Quick Revision Summary

- **Inheritance**: `class Derived : access Base { };` — IS-A relationship
- **Types**: Single, Multiple, Multilevel, Hierarchical, Hybrid (virtual)
- **`public` inheritance**: Preserves access levels (most common)
- **`protected` inheritance**: public → protected in derived
- **`private` inheritance**: Everything becomes private in derived
- **Constructor order**: Base → Member objects → Derived
- **Destructor order**: Derived → Member objects → Base (reverse)
- **Virtual inheritance**: Solves diamond problem with `virtual public`
- **Virtual destructor**: Required in base class for proper cleanup
