## 4.1 Encapsulation (Data Hiding)

**Encapsulation** is the concept of **bundling data** (member variables) and **behavior** (member functions) together inside a class, while **restricting direct access** to some of the object's internal data.

### Key Idea

> Keep your data **private**, expose only what is necessary through a **public interface**.

### Why Encapsulation Matters

| Benefit | Explanation |
|---|---|
| **Security** | Prevents accidental or malicious modification of data |
| **Maintainability** | Internal implementation can change without affecting external code |
| **Validation** | Controls what values can be assigned to data members |
| **Debugging** | Easier to trace data changes through controlled access points |

### Without Encapsulation (Bad Practice)

```c++
class BankAccount {
    public:
        double balance;  // Public — anyone can modify directly
};

int main() {
    BankAccount acc;
    acc.balance = -1000;  // No validation! Negative balance allowed
    return 0;
}
```

### With Encapsulation (Good Practice)

```c++
class BankAccount {
    private:
        double balance;  // Private — hidden from outside
    
    public:
        BankAccount() : balance(0) {}
        
        void deposit(double amount) {
            if (amount > 0) {
                balance += amount;
            }
        }
        
        bool withdraw(double amount) {
            if (amount > 0 && amount <= balance) {
                balance -= amount;
                return true;
            }
            return false;
        }
        
        double getBalance() const {  // Read-only access
            return balance;
        }
};

int main() {
    BankAccount acc;
    // acc.balance = -1000;  // ERROR! private member
    acc.deposit(500);        // Use public interface instead
    cout << "Balance: " << acc.getBalance() << endl;
    return 0;
}
```

## 4.2 Getters and Setters

**Getters** (accessors) and **Setters** (mutators) provide controlled access to private members.

### Standard Pattern

```c++
class Student {
    private:
        string name;
        int age;
        double cgpa;
    
    public:
        // Getter — read-only access
        string getName() const { return name; }
        int getAge() const { return age; }
        double getCgpa() const { return cgpa; }
        
        // Setter — write access with validation
        void setName(const string& n) { name = n; }
        
        void setAge(int a) {
            if (a > 0 && a < 150) {
                age = a;
            } else {
                cout << "Invalid age!" << endl;
            }
        }
        
        void setCgpa(double c) {
            if (c >= 0.0 && c <= 10.0) {
                cgpa = c;
            } else {
                cout << "Invalid CGPA! Must be between 0.0 and 10.0" << endl;
            }
        }
};
```

### Full Example with Validation

```c++
#include <iostream>
#include <string>
using namespace std;

class Employee {
    private:
        int id;
        string name;
        double salary;
        string department;
    
    public:
        // Constructor
        Employee(int i, string n, double s, string d) {
            id = i;
            setName(n);     // Use setter for validation
            setSalary(s);
            department = d;
        }
        
        // Getters
        int getId() const { return id; }
        string getName() const { return name; }
        double getSalary() const { return salary; }
        string getDepartment() const { return department; }
        
        // Setters with validation
        void setName(const string& n) {
            if (n.length() >= 2) {
                name = n;
            } else {
                cout << "Name must be at least 2 characters!" << endl;
                name = "Unknown";
            }
        }
        
        void setSalary(double s) {
            if (s >= 0) {
                salary = s;
            } else {
                cout << "Salary cannot be negative!" << endl;
                salary = 0;
            }
        }
        
        void setDepartment(const string& d) {
            department = d;
        }
        
        // Business logic using private data
        void giveRaise(double percentage) {
            if (percentage > 0) {
                salary += salary * (percentage / 100);
                cout << name << " received a " << percentage 
                     << "% raise. New salary: $" << salary << endl;
            }
        }
        
        void display() const {
            cout << "ID: " << id << " | " << name 
                 << " | $" << salary << " | " << department << endl;
        }
};

int main() {
    Employee emp(101, "Arjun", 50000, "Engineering");
    emp.display();
    
    emp.giveRaise(10);  // 10% raise
    emp.setSalary(-100); // Invalid — won't set
    
    cout << "\nFinal details:" << endl;
    cout << "Name: " << emp.getName() << endl;
    cout << "Salary: $" << emp.getSalary() << endl;
    
    return 0;
}
```

**Output:**

```c++
ID: 101 | Arjun | $50000 | Engineering
Arjun received a 10% raise. New salary: $55000
Salary cannot be negative!

Final details:
Name: Arjun
Salary: $55000
```

## 4.3 Abstraction Concepts

**Abstraction** means hiding complex implementation details and showing only the essential features of an object.

### Abstract Classes vs Interfaces (in C++)

In C++, **interfaces** are implemented as **abstract classes** with **all pure virtual functions**.

| Feature | Abstract Class | Interface (C++ idiom) |
|---|---|---|
| **Pure virtual functions** | At least one | All are pure virtual |
| **Can have data members** | Yes | Typically no |
| **Can have implemented methods** | Yes | No (all pure virtual) |
| **Multiple inheritance** | Yes (with care) | Yes (safe) |
| **Keyword** | `virtual ... = 0` | `virtual ... = 0` |

### Pure Virtual Functions & Abstract Classes

```c++
#include <iostream>
#include <string>
using namespace std;

// Abstract class (cannot be instantiated)
class Shape {
    protected:
        string color;
    
    public:
        Shape(string c) : color(c) {}
        
        // Pure virtual function — must be overridden by derived classes
        virtual double area() const = 0;
        
        // Pure virtual function
        virtual void draw() const = 0;
        
        // Concrete method — shared by all derived classes
        string getColor() const { return color; }
        
        // Virtual destructor — needed for proper cleanup
        virtual ~Shape() {}
};

// Concrete derived class
class Circle : public Shape {
    private:
        double radius;
    
    public:
        Circle(string c, double r) : Shape(c), radius(r) {}
        
        double area() const override {
            return 3.14159 * radius * radius;
        }
        
        void draw() const override {
            cout << "Drawing a " << color << " circle (area: " << area() << ")" << endl;
        }
};

class Rectangle : public Shape {
    private:
        double length;
        double width;
    
    public:
        Rectangle(string c, double l, double w) : Shape(c), length(l), width(w) {}
        
        double area() const override {
            return length * width;
        }
        
        void draw() const override {
            cout << "Drawing a " << color << " rectangle (area: " << area() << ")" << endl;
        }
};

int main() {
    // Shape s("red");  // ERROR! Cannot instantiate abstract class
    
    Shape* shapes[2];
    shapes[0] = new Circle("Red", 5.0);
    shapes[1] = new Rectangle("Blue", 4.0, 6.0);
    
    for (int i = 0; i < 2; i++) {
        shapes[i]->draw();
    }
    
    // Cleanup
    for (int i = 0; i < 2; i++) {
        delete shapes[i];
    }
    
    return 0;
}
```

**Output:**

```c++
Drawing a Red circle (area: 78.5397)
Drawing a Blue rectangle (area: 24)
```

## 4.4 Friend Function and Friend Class

A **friend** is a function or class that is granted **access to the private and protected members** of another class.

### Friend Function

```c++
#include <iostream>
using namespace std;

class BankAccount {
    private:
        double balance;
        string accountNumber;
    
    public:
        BankAccount(string accNo, double bal) 
            : accountNumber(accNo), balance(bal) {}
        
        // Declare an external function as a friend
        friend void displayAccountInfo(const BankAccount& acc);
        
        // Declare another class's method as friend
        friend class Auditor;
};

// Friend function definition — can access private members
void displayAccountInfo(const BankAccount& acc) {
    cout << "Account: " << acc.accountNumber << endl;
    cout << "Balance: $" << acc.balance << endl;
    // Can access private members because it's a friend
}

class Auditor {
    public:
        void audit(const BankAccount& acc) {
            cout << "AUDIT REPORT" << endl;
            cout << "Account: " << acc.accountNumber << endl;
            cout << "Verified Balance: $" << acc.balance << endl;
        }
};

int main() {
    BankAccount acc("ACC-1001", 50000);
    displayAccountInfo(acc);  // Friend function
    
    cout << endl;
    Auditor auditor;
    auditor.audit(acc);  // Friend class
    
    return 0;
}
```

**Output:**

```c++
Account: ACC-1001
Balance: $50000

AUDIT REPORT
Account: ACC-1001
Verified Balance: $50000
```

### When to Use Friends

| Use Friend When... | Don't Use Friend When... |
|---|---|
| Operator overloading (e.g., `<<`, `>>`) | Normal getter/setter would suffice |
| Two classes are tightly coupled | It breaks encapsulation unnecessarily |
| External function needs deep access | The design can be improved with proper interfaces |

> **Caution:** Friend functions break encapsulation. Use them sparingly and only when necessary.

## 4.5 Basics of Pointers

Pointers are variables that store **memory addresses** rather than actual values.

### Pointer Declaration and Operations

```c++
#include <iostream>
using namespace std;

int main() {
    int x = 42;
    
    // Pointer declaration
    int* ptr;          // Declare a pointer to an integer
    
    // Address-of operator (&)
    ptr = &x;          // ptr now stores the address of x
    
    cout << "Value of x: " << x << endl;
    cout << "Address of x: " << &x << endl;
    cout << "Value of ptr (address it holds): " << ptr << endl;
    
    // Dereference operator (*)
    cout << "Value at address ptr: " << *ptr << endl;
    
    // Modify value through pointer
    *ptr = 100;
    cout << "x after *ptr = 100: " << x << endl;
    
    // Null pointer
    int* nullPtr = nullptr;
    // cout << *nullPtr;  // CRASH! Dereferencing null pointer
    
    return 0;
}
```

**Output:**

```c++
Value of x: 42
Address of x: 0x7ffdf3a1b4c4
Value of ptr (address it holds): 0x7ffdf3a1b4c4
Value at address ptr: 42
x after *ptr = 100: 100
```

### Pointer Types

```c++
int main() {
    int x = 10;
    double y = 3.14;
    char c = 'A';
    
    int* intPtr = &x;         // Pointer to int
    double* doublePtr = &y;   // Pointer to double
    char* charPtr = &c;       // Pointer to char
    
    cout << "Size of int pointer: " << sizeof(intPtr) << " bytes" << endl;
    // All pointers are the same size (typically 8 bytes on 64-bit)
    cout << "Size of double pointer: " << sizeof(doublePtr) << " bytes" << endl;
    cout << "Size of char pointer: " << sizeof(charPtr) << " bytes" << endl;
    
    return 0;
}
```

### Pointers and Arrays

```c++
#include <iostream>
using namespace std;

int main() {
    int arr[5] = {10, 20, 30, 40, 50};
    
    int* ptr = arr;  // Array name acts as pointer to first element
    
    cout << "Array traversal using pointer:" << endl;
    for (int i = 0; i < 5; i++) {
        cout << "arr[" << i << "] = " << *ptr << " (address: " << ptr << ")" << endl;
        ptr++;  // Move to next element (advances by sizeof(int) bytes)
    }
    
    return 0;
}
```

**Output:**

```c++
Array traversal using pointer:
arr[0] = 10 (address: 0x7fff1234)
arr[1] = 20 (address: 0x7fff1238)
arr[2] = 30 (address: 0x7fff123c)
arr[3] = 40 (address: 0x7fff1240)
arr[4] = 50 (address: 0x7fff1244)
```

## 4.6 The `this` Pointer

Every non-static member function has an implicit parameter called the **`this` pointer**, which points to the current object.

### Use Cases of `this`

#### 1. Disambiguation (parameter name vs member name)

```c++
class Student {
    private:
        string name;
        int age;
    
    public:
        // Without 'this' — need different parameter names
        void setName(string n) { name = n; }  // Works but verbose
        
        // With 'this' — can use same name
        void setAge(int age) {
            this->age = age;  // this->age is member; age is parameter
        }
};
```

#### 2. Method Chaining (returning `*this`)

```c++
#include <iostream>
using namespace std;

class Calculator {
    private:
        int result;
    
    public:
        Calculator() : result(0) {}
        
        Calculator& add(int x) {
            result += x;
            return *this;  // Return reference to current object
        }
        
        Calculator& subtract(int x) {
            result -= x;
            return *this;
        }
        
        Calculator& multiply(int x) {
            result *= x;
            return *this;
        }
        
        void display() {
            cout << "Result: " << result << endl;
        }
};

int main() {
    Calculator calc;
    
    // Method chaining — each method returns *this
    calc.add(10).subtract(3).multiply(4);
    calc.display();  // (10 - 3) * 4 = 28
    
    return 0;
}
```

**Output:**

```c++
Result: 28
```

#### 3. Preventing Self-Assignment

```c++
class String {
    private:
        char* data;
    
    public:
        String& operator=(const String& other) {
            if (this != &other) {  // Self-assignment check
                delete[] data;     // Release old resource
                // ... allocate and copy new data ...
            }
            return *this;
        }
};
```

## Coding Examples

### Example 1 — Complete Encapsulated Class

```c++
#include <iostream>
#include <string>
using namespace std;

class Temperature {
    private:
        double celsius;
        
        // Helper (private) — not exposed to users
        double toFahrenheit() const {
            return (celsius * 9.0 / 5.0) + 32;
        }
        
        double toKelvin() const {
            return celsius + 273.15;
        }
    
    public:
        Temperature() : celsius(0) {}
        
        void setCelsius(double c) {
            if (c >= -273.15) {  // Absolute zero check
                celsius = c;
            } else {
                cout << "Temperature cannot be below absolute zero!" << endl;
                celsius = -273.15;
            }
        }
        
        void setFahrenheit(double f) {
            double c = (f - 32) * 5.0 / 9.0;
            setCelsius(c);  // Reuse validation
        }
        
        double getCelsius() const { return celsius; }
        double getFahrenheit() const { return toFahrenheit(); }
        double getKelvin() const { return toKelvin(); }
        
        void display() const {
            cout << celsius << "°C = " 
                 << toFahrenheit() << "°F = " 
                 << toKelvin() << "K" << endl;
        }
};

int main() {
    Temperature temp;
    temp.setCelsius(25);
    cout << "25°C in different units:" << endl;
    temp.display();
    
    temp.setFahrenheit(98.6);
    cout << "\n98.6°F in different units:" << endl;
    temp.display();
    
    temp.setCelsius(-500);  // Invalid — below absolute zero
    cout << "\nAfter invalid input:" << endl;
    temp.display();
    
    return 0;
}
```

**Output:**

```c++
25°C in different units:
25°C = 77°F = 298.15K

98.6°F in different units:
37°C = 98.6°F = 310.15K

Temperature cannot be below absolute zero!

After invalid input:
-273.15°C = -459.67°F = 0K
```

### Example 2 — this Pointer for Method Chaining

```c++
#include <iostream>
#include <string>
using namespace std;

class Message {
    private:
        string text;
    
    public:
        Message() : text("") {}
        
        Message& append(const string& str) {
            text += str;
            return *this;
        }
        
        Message& prepend(const string& str) {
            text = str + text;
            return *this;
        }
        
        Message& toUpperCase() {
            for (char& c : text) {
                if (c >= 'a' && c <= 'z')
                    c = c - 'a' + 'A';
            }
            return *this;
        }
        
        Message& addNewline() {
            text += '\n';
            return *this;
        }
        
        void print() const {
            cout << text << endl;
        }
};

int main() {
    Message msg;
    
    msg.append("Hello")
       .append(" ")
       .append("World")
       .addNewline()
       .append("This is method chaining!")
       .toUpperCase();
    
    msg.print();
    return 0;
}
```

**Output:**

```c++
HELLO WORLD
THIS IS METHOD CHAINING!
```

### Example 3 — Friend Function for Operator Overloading (<<)

```c++
#include <iostream>
#include <string>
using namespace std;

class Vector2D {
    private:
        double x;
        double y;
    
    public:
        Vector2D(double xVal = 0, double yVal = 0) : x(xVal), y(yVal) {}
        
        // Friend function for output operator
        friend ostream& operator<<(ostream& os, const Vector2D& v);
        
        // Friend function for addition
        friend Vector2D operator+(const Vector2D& a, const Vector2D& b);
};

// Friend function definitions
ostream& operator<<(ostream& os, const Vector2D& v) {
    os << "(" << v.x << ", " << v.y << ")";
    return os;
}

Vector2D operator+(const Vector2D& a, const Vector2D& b) {
    return Vector2D(a.x + b.x, a.y + b.y);
}

int main() {
    Vector2D v1(3.0, 4.0);
    Vector2D v2(1.5, 2.5);
    Vector2D v3 = v1 + v2;
    
    cout << "v1 = " << v1 << endl;
    cout << "v2 = " << v2 << endl;
    cout << "v1 + v2 = " << v3 << endl;
    
    return 0;
}
```

**Output:**

```c++
v1 = (3, 4)
v2 = (1.5, 2.5)
v1 + v2 = (4.5, 6.5)
```

## Common Mistakes

| Mistake | Wrong | Correct |
|---|---|---|
| **Making all members public** | No encapsulation | Mark data `private`, expose via getters/setters |
| **No validation in setters** | `setAge(int a) { age = a; }` | `if (a > 0) age = a;` |
| **Forgetting `const` in getters** | `int getX() { return x; }` | `int getX() const { return x; }` |
| **Overusing friend functions** | Friends for every accessor | Use getters/setters first; friends when necessary |
| **Dereferencing nullptr** | `*ptr` when `ptr = nullptr` | Check `if (ptr != nullptr)` first |
| **Forgetting `*this` return type** | `void add(int x) { ... }` returns nothing | `ClassName& add(int x) { ... return *this; }` |

## Interview Questions

1. **What is encapsulation and why is it important?**
   - Bundling data + methods, restricting direct access. Important for **security**, **maintainability**, **validation**, and **debugging**.

2. **What is the difference between abstraction and encapsulation?**
   - **Encapsulation** hides the *internal state* (data hiding). **Abstraction** hides the *implementation details* (showing only essential features). Encapsulation is *how* you hide data; abstraction is *what* you expose to the user.

3. **What is a friend function? When would you use one?**
   - A non-member function granted access to private/protected members. Used for operator overloading (especially `<<` and `>>`) and when external functions need deep class access.

4. **What is the `this` pointer?**
   - An implicit pointer available in all non-static member functions that points to the current object. Used for disambiguation, method chaining, and self-assignment checks.

5. **Can an abstract class have a constructor?**
   - Yes. Even though you can't instantiate an abstract class directly, its constructor is called when a derived class object is created (via constructor chaining).

## Quick Revision Summary

- **Encapsulation**: Bundle data + behavior; hide internal state
- **Private**: Accessible only within the class
- **Getters**: Provide read-only access (`const` methods)
- **Setters**: Provide write access with **validation**
- **Abstraction**: Hide complexity; show only essential features
- **Pure virtual function**: `virtual void func() = 0;` — must be overridden
- **Abstract class**: Has at least one pure virtual function (cannot be instantiated)
- **Friend function/class**: Granted access to private members (use sparingly)
- **Pointer**: Stores memory address (`*` to declare/dereference; `&` to get address)
- **`this` pointer**: Implicit pointer to current object; enables method chaining
