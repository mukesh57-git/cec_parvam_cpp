## 3.1 What is a Constructor?

A **constructor** is a special member function that is **automatically called** when an object is created. It is used to initialize the object's data members.

### Key Characteristics

- Same name as the class
- **No return type** (not even `void`)
- Automatically invoked when an object is instantiated
- Can be **overloaded** (multiple constructors with different parameters)
- If not defined, the compiler provides a **default constructor** (no parameters, does nothing)

### Basic Syntax

```c++
class ClassName {
    public:
        ClassName() {
            // Constructor body — initialization code
        }
};
```

---

## 3.2 Types of Constructors

### 1. Default Constructor

A constructor that takes **no parameters**. If you don't define any constructor, the compiler generates one automatically.

```c++
#include <iostream>
using namespace std;

class Student {
    public:
        string name;
        int age;
        
        // Default constructor
        Student() {
            name = "Unknown";
            age = 0;
            cout << "Default constructor called!" << endl;
        }
        
        void display() {
            cout << name << " (" << age << ")" << endl;
        }
};

int main() {
    Student s1;      // Default constructor called
    s1.display();
    
    Student s2();    // OOPS! This is a function declaration, not an object!
    // Student s2;   // Correct way to call default constructor
    return 0;
}
```

**Output:**

```c++
Default constructor called!
Unknown (0)
```

> **Warning:** `Student s2();` is interpreted as a **function prototype** returning a `Student` object — NOT an object declaration. Use `Student s2;` instead.

### 2. Parameterized Constructor

A constructor that takes **arguments** to initialize an object with specific values.

```c++
#include <iostream>
#include <string>
using namespace std;

class Student {
    public:
        string name;
        int age;
        double cgpa;
        
        // Parameterized constructor
        Student(string n, int a, double c) {
            name = n;
            age = a;
            cgpa = c;
            cout << "Parameterized constructor called for " << name << endl;
        }
        
        void display() {
            cout << name << " | Age: " << age << " | CGPA: " << cgpa << endl;
        }
};

int main() {
    // Two ways to call parameterized constructor
    Student s1("Arjun", 20, 8.75);           // Implicit call
    Student s2 = Student("Priya", 21, 9.2);  // Explicit call
    
    s1.display();
    s2.display();
    return 0;
}
```

**Output:**

```c++
Parameterized constructor called for Arjun
Parameterized constructor called for Priya
Arjun | Age: 20 | CGPA: 8.75
Priya | Age: 21 | CGPA: 9.2
```

### 3. Copy Constructor

A constructor that creates a new object by **copying** an existing object of the same type.

```c++
#include <iostream>
#include <cstring>
using namespace std;

class Student {
    public:
        char* name;
        int age;
        
        // Parameterized constructor
        Student(const char* n, int a) {
            age = a;
            name = new char[strlen(n) + 1];
            strcpy(name, n);
        }
        
        // Copy constructor (deep copy)
        Student(const Student& s) {
            age = s.age;
            name = new char[strlen(s.name) + 1];
            strcpy(name, s.name);
            cout << "Copy constructor called!" << endl;
        }
        
        void display() {
            cout << name << " (" << age << ")" << endl;
        }
        
        ~Student() {
            delete[] name;  // Clean up heap memory
        }
};

int main() {
    Student s1("Arjun", 20);
    Student s2 = s1;       // Copy constructor called
    
    s1.display();
    s2.display();
    
    // Prove it's a deep copy — modifying s2 doesn't affect s1
    s2.name[0] = 'P';
    cout << "\nAfter modification:" << endl;
    s1.display();  // Still "Arjun"
    s2.display();  // Now "Prjun"
    
    return 0;
}
```

**Output:**

```c++
Copy constructor called!
Arjun (20)
Arjun (20)

After modification:
Arjun (20)
Prjun (20)
```

#### Shallow vs Deep Copy

| Aspect | Shallow Copy | Deep Copy |
|---|---|---|
| **What is copied?** | Only pointer addresses | The actual data being pointed to |
| **Default copy constructor?** | Yes (shallow) | No (must define manually) |
| **Memory sharing?** | Both objects share same memory | Each object has its own memory |
| **Risk** | Double deletion, dangling pointers | None (if implemented correctly) |
| **When needed?** | No dynamic memory | When class has pointer members |

**Shallow Copy Problem:**

```c++
class Shallow {
    public:
        int* data;
        Shallow(int val) {
            data = new int(val);
        }
        // No copy constructor → uses default shallow copy
};

int main() {
    Shallow a(10);
    Shallow b = a;    // Shallow copy — both point to SAME memory
    delete a.data;    // Frees the memory
    // b.data now is a DANGLING POINTER — accessing it is undefined behavior!
}
```

## 3.3 Constructor Overloading

Like regular functions, constructors can be **overloaded** — multiple constructors with different parameter lists.

```c++
#include <iostream>
#include <string>
using namespace std;

class Rectangle {
    private:
        double length;
        double width;
    
    public:
        // 1. Default constructor
        Rectangle() {
            length = 0;
            width = 0;
            cout << "Default constructor" << endl;
        }
        
        // 2. Parameterized constructor (1 parameter — square)
        Rectangle(double side) {
            length = side;
            width = side;
            cout << "Square constructor" << endl;
        }
        
        // 3. Parameterized constructor (2 parameters)
        Rectangle(double l, double w) {
            length = l;
            width = w;
            cout << "Two-parameter constructor" << endl;
        }
        
        // 4. Copy constructor
        Rectangle(const Rectangle& r) {
            length = r.length;
            width = r.width;
            cout << "Copy constructor" << endl;
        }
        
        double area() { return length * width; }
};

int main() {
    Rectangle r1;              // Calls (1) Default
    Rectangle r2(5);           // Calls (2) Single parameter
    Rectangle r3(4, 6);        // Calls (3) Two parameters
    Rectangle r4 = r3;         // Calls (4) Copy constructor
    
    cout << "\nAreas: " << r1.area() << ", " << r2.area() 
         << ", " << r3.area() << ", " << r4.area() << endl;
    
    return 0;
}
```

**Output:**

```c++
Default constructor
Square constructor
Two-parameter constructor
Copy constructor

Areas: 0, 25, 24, 24
```

### Constructor Initialization List (Preferred Syntax)

Instead of assigning values inside the constructor body, you can use an **initialization list**:

```c++
class Example {
    private:
        int a;
        int b;
        const int c;        // const members MUST be initialized in init list
        int& ref;           // reference members MUST be initialized in init list
    
    public:
        // Initialization list syntax
        Example(int x, int y, int z) : a(x), b(y), c(z), ref(a) {
            // Body is now empty (or has extra logic)
        }
};
```

**Why initialization lists are preferred:**
1. **Required** for `const` and reference members
2. **More efficient** — members are initialized directly instead of being default-constructed then assigned
3. **Cleaner** syntax for base class constructor calls (inheritance)

## 3.4 Destructor Basics

A **destructor** is a special member function that is **automatically called** when an object is destroyed (goes out of scope or is deleted).

### Syntax

```c++
class ClassName {
    public:
        ~ClassName() {
            // Cleanup code (release memory, close files, etc.)
        }
};
```

### Key Characteristics

- Same name as class, **prefixed with `~`** (tilde)
- **No return type**
- **No parameters** (cannot be overloaded — exactly one destructor per class)
- Automatically called when:
  - Stack object goes out of scope
  - Heap object is `delete`d
  - Program ends (for global/static objects)

### Example

```c++
#include <iostream>
using namespace std;

class Resource {
    private:
        int* data;
        int id;
    
    public:
        Resource(int i, int val) : id(i) {
            data = new int(val);
            cout << "Resource " << id << " acquired (value: " << *data << ")" << endl;
        }
        
        ~Resource() {
            delete data;  // Release heap memory
            cout << "Resource " << id << " released" << endl;
        }
        
        void show() {
            cout << "Resource " << id << ": " << *data << endl;
        }
};

void function() {
    Resource local(2, 200);
    local.show();
    // local destroyed automatically when function ends
}

int main() {
    Resource r1(1, 100);
    r1.show();
    
    function();
    
    Resource* r3 = new Resource(3, 300);
    r3->show();
    delete r3;  // Explicit destruction
    
    cout << "End of main" << endl;
    // r1 destroyed automatically
    return 0;
}
```

**Output:**

```c++
Resource 1 acquired (value: 100)
Resource 1: 100
Resource 2 acquired (value: 200)
Resource 2: 200
Resource 2 released
Resource 3 acquired (value: 300)    
Resource 3: 300
Resource 3 released
End of main
Resource 1 released
```

## 3.5 Role in Memory Management

### RAII (Resource Acquisition Is Initialization)

RAII is a core C++ programming idiom where:
- **Resources** (memory, files, locks, network connections) are **acquired** in the constructor
- **Released** in the destructor
- This ensures resources are properly freed even when exceptions occur

### Without RAII (prone to leaks):

```c++
void badFunction() {
    int* data = new int[1000];
    // ... do some work ...
    if (error) return;       // OOPS! Memory leak — forgot to delete
    delete[] data;
}
```

### With RAII:

```c++
class IntArray {
    private:
        int* arr;
        int size;
    
    public:
        IntArray(int s) : size(s) {
            arr = new int[s];  // Resource acquired in constructor
            cout << "Array of " << size << " elements allocated" << endl;
        }
        
        ~IntArray() {
            delete[] arr;      // Resource released in destructor
            cout << "Array deallocated" << endl;
        }
        
        void set(int index, int value) { arr[index] = value; }
        int get(int index) { return arr[index]; }
};

void safeFunction() {
    IntArray nums(100);    // Constructor allocates
    nums.set(0, 42);
    // ... do some work ...
    if (error) return;     // Destructor automatically cleans up!
    // Destructor called when nums goes out of scope
}

int main() {
    safeFunction();
    // No memory leaks!
    return 0;
}
```

### Preventing Memory Leaks — Best Practices

```c++
#include <iostream>
#include <string>
using namespace std;

class SafeString {
    private:
        char* buffer;
        int length;
    
    public:
        // Constructor — acquire resource
        SafeString(const char* str) {
            length = strlen(str);
            buffer = new char[length + 1];
            strcpy(buffer, str);
            cout << "Memory allocated for \"" << buffer << "\"" << endl;
        }
        
        // Copy constructor — deep copy
        SafeString(const SafeString& other) {
            length = other.length;
            buffer = new char[length + 1];
            strcpy(buffer, other.buffer);
            cout << "Deep copy of \"" << buffer << "\"" << endl;
        }
        
        // Assignment operator (for completeness — Rule of Three)
        SafeString& operator=(const SafeString& other) {
            if (this != &other) {          // Self-assignment check
                delete[] buffer;            // Release old memory
                length = other.length;
                buffer = new char[length + 1];
                strcpy(buffer, other.buffer);
            }
            return *this;
        }
        
        // Destructor — release resource
        ~SafeString() {
            delete[] buffer;
            cout << "Memory deallocated" << endl;
        }
        
        void display() { cout << buffer << endl; }
};

int main() {
    SafeString s1("Hello, C++!");
    SafeString s2 = s1;   // Deep copy via copy constructor
    
    s1.display();
    s2.display();
    
    return 0;
    // Both s1 and s2 clean up their OWN memory
}
```

**Output:**

```c++
Memory allocated for "Hello, C++!"
Deep copy of "Hello, C++!"
Hello, C++!
Hello, C++!
Memory deallocated
Memory deallocated
```

## Coding Examples

### Example 1 — Complete Class with Full Constructor Suite

```c++
#include <iostream>
#include <cstring>
using namespace std;

class Person {
    private:
        char* name;
        int age;
    
    public:
        // Default constructor
        Person() {
            name = new char[1];
            name[0] = '\0';
            age = 0;
            cout << "Default constructor" << endl;
        }
        
        // Parameterized constructor
        Person(const char* n, int a) {
            name = new char[strlen(n) + 1];
            strcpy(name, n);
            age = a;
            cout << "Parameterized constructor: " << name << endl;
        }
        
        // Copy constructor
        Person(const Person& p) {
            name = new char[strlen(p.name) + 1];
            strcpy(name, p.name);
            age = p.age;
            cout << "Copy constructor: " << name << endl;
        }
        
        // Destructor
        ~Person() {
            cout << "Destructor: " << name << endl;
            delete[] name;
        }
        
        void display() {
            cout << name << " (" << age << " years old)" << endl;
        }
};

int main() {
    Person p1;                            // Default
    Person p2("Arjun", 20);               // Parameterized
    Person p3 = p2;                       // Copy
    
    cout << "\n--- Person details ---" << endl;
    p1.display();
    p2.display();
    p3.display();
    
    cout << "\n--- End of main ---" << endl;
    return 0;
}
```

**Output:**

```c++
Default constructor
Parameterized constructor: Arjun
Copy constructor: Arjun

--- Person details ---
 (0 years old)
Arjun (20 years old)
Arjun (20 years old)

--- End of main ---
Destructor: Arjun
Destructor: Arjun
Destructor:
```

### Example 2 — Constructor Initialization List

```c++
#include <iostream>
using namespace std;

class Point {
    private:
        const int x;    // const member — MUST use init list
        const int y;    // const member — MUST use init list
    
    public:
        // Initialization list — only way to set const members
        Point(int xVal, int yVal) : x(xVal), y(yVal) {
            // x = xVal;  // ERROR! Can't assign to const
            cout << "Point(" << x << ", " << y << ") created" << endl;
        }
        
        void display() const {
            cout << "(" << x << ", " << y << ")" << endl;
        }
};

int main() {
    Point p1(10, 20);
    Point p2(-5, 15);
    
    p1.display();
    p2.display();
    
    return 0;
}
```

**Output:**

```c++
Point(10, 20) created
Point(-5, 15) created
(10, 20)
(-5, 15)
```

### Example 3 — Array of Objects with Constructors

```c++
#include <iostream>
#include <string>
using namespace std;

class Product {
    private:
        int id;
        string name;
        double price;
    
    public:
        Product() : id(0), name("None"), price(0.0) {}
        
        Product(int i, string n, double p) : id(i), name(n), price(p) {}
        
        void display() {
            cout << "ID: " << id << " | " << name << " | $" << price << endl;
        }
};

int main() {
    // Stack array — default constructor for each element
    Product inventory[3];
    
    cout << "Inventory (default initialized):" << endl;
    for (int i = 0; i < 3; i++)
        inventory[i].display();
    
    // Dynamic array — default constructor for each element
    Product* heapInventory = new Product[3];
    
    // Re-assign with custom values
    heapInventory[0] = Product(101, "Laptop", 999.99);
    heapInventory[1] = Product(102, "Mouse", 29.99);
    heapInventory[2] = Product(103, "Keyboard", 79.99);
    
    cout << "\nHeap Inventory (assigned):" << endl;
    for (int i = 0; i < 3; i++)
        heapInventory[i].display();
    
    delete[] heapInventory;
    return 0;
}
```

**Output:**

```c++
Inventory (default initialized):
ID: 0 | None | $0
ID: 0 | None | $0
ID: 0 | None | $0

Heap Inventory (assigned):
ID: 101 | Laptop | $999.99
ID: 102 | Mouse | $29.99
ID: 103 | Keyboard | $79.99
```

## Common Mistakes

| Mistake | Wrong | Correct |
|---|---|---|
| **Calling default constructor with parentheses** | `Student s();` (function declaration!) | `Student s;` |
| **Forgetting to define copy constructor** | Default shallow copy corrupts heap data | Implement deep copy constructor |
| **Constructor body instead of init list for const** | `Point(int x) { this->x = x; }` | `Point(int x) : x(x) {}` |
| **Destructor without `delete[]`** | `delete buffer;` instead of `delete[] buffer;` | Use `delete[]` for array allocations |
| **Self-assignment not checked** | `a = a;` crashes in operator= | Check `if (this != &other)` |
| **No virtual destructor in base class** | Deleting derived via base pointer — undefined behavior | Declare `virtual ~Base()` |

## Interview Questions

1. **What is a constructor? What are its types?**
   - A special member function auto-called on object creation. Types: **Default** (no params), **Parameterized** (with params), **Copy** (creates from existing object).

2. **What is the difference between shallow copy and deep copy?**
   - **Shallow copy**: Copies pointer addresses — both objects share same memory (default behavior). **Deep copy**: Copies actual data — each object has independent memory.

3. **What is a destructor and when is it called?**
   - A special member function `~ClassName()` auto-called when object is destroyed (scope exit, `delete`, program end). Used for cleanup.

4. **Explain RAII in C++.**
   - Resource Acquisition Is Initialization: Resources are acquired in constructors and released in destructors. Ensures exception-safe resource management.

5. **What is the Rule of Three?**
   - If a class defines any of: **destructor**, **copy constructor**, or **copy assignment operator**, it should define all three (typically due to dynamic memory management).

## Quick Revision Summary

- **Constructor**: Same name as class, no return type, auto-called on object creation
- **Default constructor**: No parameters; compiler generates one if none defined
- **Parameterized constructor**: Takes arguments for custom initialization
- **Copy constructor**: `ClassName(const ClassName& other)` — deep copy when using pointers
- **Destructor**: `~ClassName()` — no parameters, no overloading, auto-called on destruction
- **Initialization list**: `: member1(val1), member2(val2){}` — required for `const` and `&` members
- **RAII**: Acquire resources in constructor, release in destructor
- **Shallow copy** = copies pointers; **Deep copy** = copies pointed-to data
- **Rule of Three**: Destructor + Copy Constructor + Copy Assignment Operator
